```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: Material properties (e.g., Density, Hardness, and Thermal Conductivity)

# Load the CSV file
file_path = "https://raw.githubusercontent.com/MonikSense/MonikSense/refs/heads/main/rgosno4%20final.CSV"  # Replace with your CSV file path
data = pd.read_csv(file_path)
#data = {
    #"Density (g/cm)": [7.8, 2.7, 8.9, 19.3, 4.5],
   # "Hardness (Mohs)": [4.0, 2.75, 3.5, 6.0, 6.0],
  #  "Thermal Conductivity (W/mK)": [80, 237, 401, 173, 22]
#}

# Convert to DataFrame
df = pd.DataFrame(data)

# Compute Pearson correlation matrix
correlation_matrix = df.corr(method="pearson")

# Create a heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.8)

plt.xticks(rotation=0)
plt.xticks(fontsize=13, fontweight= 'bold')
plt.yticks(rotation=90)
plt.yticks(fontsize=13, fontweight= 'bold')

# Set title
plt.title("Pearson Correlation Heatmap of Sensing Parameters", fontsize=14, fontweight= 'bold')

plt.show()
```
